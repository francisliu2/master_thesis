```python
data_to_go = np.nan_to_num(data_to_go) #Replace NaN with zero and infinity with large finite numbers.
```

```python
# preprocessing of the data
# 1. price relative vector, element-wise division (done)
PRV = data_to_go[:,1:,:]/data_to_go[:,:-1,:]
PRV = np.nan_to_num(PRV) 
# 2. change in total portfolio value in a period
Pt = np.nan_to_num(PRV[0,2,:]/PRV[0,1,:]).mean() - 1
# 3. logarithmic rate of return
pt = math.log(Pt + 1,math.exp(1))
# 4. Transaction Cost
```
